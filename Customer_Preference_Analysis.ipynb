```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv") 
df
```

```python
display(df.columns)

df['Cuisines'] = df['Cuisines'].astype(str).apply(lambda x: x.split(',')[0]) 
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(14, 6))
top_cuisines = df['Cuisines'].value_counts().head(15).index
filtered_df = df[df['Cuisines'].isin(top_cuisines)]
sns.boxplot(data=filtered_df, x='Cuisines', y='Aggregate rating', palette='coolwarm')
plt.title('Relationship Between the Type of Cuisine and the Restaurant\'s Rating')
plt.xlabel('Cuisine')
plt.ylabel('Aggregate Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
popular_cuisines = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=popular_cuisines.values, y=popular_cuisines.index, palette='magma')
plt.title('Most Popular Cuisines by Number of Votes')
plt.xlabel('Total Votes')
plt.ylabel('Cuisine')
plt.show()
```

```python
top_rated_cuisines = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_rated_cuisines.values, y=top_rated_cuisines.index, palette='viridis')
plt.title('Top 10 Cuisines with Higher Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')
plt.show()
```
